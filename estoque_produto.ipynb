1 - Crie um dicion√°rio chamado estoque.
2 - O usu√°rio poder√° cadastrar v√°rios produtos, fornecendo:
C√≥digo do produto (ex: "P001")
Nome do produto (ex: "Teclado")
Pre√ßo (float)
Quantidade em estoque (int)
3 - Ap√≥s os cadastros, o sistema deve permitir:
Listar todos os produtos cadastrados.
Consultar os dados de um produto pelo c√≥digo.
Atualizar o estoque (aumentar ou diminuir a quantidade).

```python
estoque = {}

def cadastrar_produto(estoque_loja):
    print('--- Cadastro de Novo Produto ---')

    codigo = input('Informe o c√≥digo do novo Produto (ex: P001): ').upper()

    if codigo in estoque_loja:
        print(f'‚ùå Erro: O c√≥digo {codigo} j√° exixte no cadastro. Tente novamente! ')
        return

    nome = input('Digite o nome do produto: ')

    try:
        preco = float(input('Digite o pre√ßo do produto: R$ '))
        quantidade = int(input('Informe a quantidade em estoque: '))

    except ValueError:
        print('‚ùå Erro: O pre√ßo e a quantidade devem ser valores n√∫mericos. Opera√ß√£o cancelada!')
        return

    estoque_loja[codigo] = {
        "nome" : nome,
        "preco" : preco,
        "quantidade" : quantidade
    }

    print(f'\n‚úÖ Produto {nome}. Cadastrado com sucesso!')

def listar_produtos(estoque_loja):
    print('\n --- Listar todos os produtos da Loja ---')

    if not estoque_loja:
        print('Nenhum produto cadastrado no estoque.')
        return

    for codigo, dados in estoque_loja.items():
        print(f'Codigo: {codigo}')
        print(f'Nome: {dados["nome"]}')
        print(f'Pre√ßo: R$ {dados["preco"]:.2f}')
        print(f'Quantidade: {dados["quantidade"]} unidades.')
        print('-' * 25)

def consultar_produto(estoque_loja):
    print('\n --- Consulta de Produto por c√≥digo ---')

    codigo = input('Informe o c√≥digo do produto que deseja consultar: ').upper()

    if codigo in estoque_loja:
        dados = estoque_loja[codigo]
        print('\n --- Dados do Produto ---')
        print(f'Codigo: {codigo}')
        print(f'Nome: {dados["nome"]}')
        print(f'Preco: R$ {dados["preco"]:.2f}')
        print(f'Quantidade: {dados["quantidade"]} unidades.')

    else:
        print(f'‚ùå Erro: Produto com c√≥digo "{codigo}" n√£o encontrado.')

def atualizar_estoque(estoque_loja):
    print('--- Atualiza√ß√£o de Estoque ---')

    codigo = input('Informe o c√≥digo do produto que deseja atualizar: ').upper()

    if codigo in estoque_loja:
        try:
            print(f'Quantidade atual de {estoque_loja[codigo]["nome"]} : {estoque_loja[codigo]["quantidade"]}')

            alteracao = int(input('Digite a quantidade a ser adicionada (‚ûï) ou removida (‚ûñ): '))

            nova_quantidade = estoque_loja[codigo]["quantidade"] + alteracao

            if nova_quantidade < 0:
                print('‚ùå Erro: O estoque n√£o pode ficar negativo. Opera√ß√£o cancelada.')

            else:
                estoque_loja[codigo]["quantidade"] = nova_quantidade
                print(f'\n ‚úÖ Estoque do produto: "{estoque_loja[codigo]["nome"]}" atualizado para {nova_quantidade} unidades.')
        except ValueError:
            print('‚ùå Erro: A quantidade deve ser um n√∫mero Inteiro.')

    else:
        print(f'‚ùå Erro: Produto com c√≥digo {codigo} n√£o encontrado.')

def calcular_valor_total(estoque_loja):
    print('\n --- Valor total do Estoque ---')

    if not estoque_loja:
        print('‚ùå Erro: Nenhum produto cadastrado para calcular o valor.')
        return

    valor_total = 0.0
    for dados in estoque_loja.values():
        valor_total += dados["preco"] * dados["quantidade"]

    print(f'O valor total do estoque √©: R$ {valor_total:.2f} ')

while True:
    print("\n===============================")
    print('üíª  SISTEMA DE CONTROLE DE ESTOQUE   ')
    print("===============================")
    print('''
    1 - Cadastrar Produto
    2 - Listar todos os Produtos
    3 - Consultar Produto pelo C√≥digo
    4 - Atualizar quantidade em estoque
    5 - Calcular o valor total em estoque
    6 - Sair do sistema
    ''')

    opcao = input('Escolha uma op√ß√£o: ')

    match opcao:
        case '1':
            cadastrar_produto(estoque)

        case '2':
            listar_produtos(estoque)

        case '3':
            consultar_produto(estoque)

        case '4':
            atualizar_estoque(estoque)

        case '5':
            calcular_valor_total(estoque)

        case '6':
            print('\nSaindo do sistema. At√© Logo')
            break

        case _:
            print('\nOp√ß√£o Invalida! Escolha uma op√ß√£o entre os n√∫meros de 1 a 6. ')


```
